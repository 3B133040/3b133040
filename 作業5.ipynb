```python
import numpy as np
import matplotlib.pyplot as plt

def plot_quadratic_equation(a, b, c):
    x = np.linspace(-10, 10, 400)  # 在範圍[-10, 10]內產生400個點作為x座標
    y = a * x**2 + b * x + c  # 計算對應的y座標

    # 繪製圖形
    plt.plot(x, y, label='Y = {}X^2 + {}X + {}'.format(a, b, c))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Quadratic Equation')
    plt.axhline(0, color='black', linewidth=0.5)  # 繪製x軸
    plt.axvline(0, color='black', linewidth=0.5)  # 繪製y軸
    plt.grid(True, linestyle='--', alpha=0.7)

    # 找出根和極值的座標值
    roots = np.roots([a, b, c])  # 找出方程式的根
    vertex_x = -b / (2 * a)  # 找出頂點的x座標
    vertex_y = a * vertex_x**2 + b * vertex_x + c  # 找出頂點的y座標

    # 標註根和極值的座標值
    for root in roots:
        plt.scatter(root, 0, color='red', marker='o')
        plt.annotate('({}, 0)'.format(round(root, 2)), (root, 0), xytext=(root, -10),
                     textcoords='offset points', ha='center')

    plt.scatter(vertex_x, vertex_y, color='blue', marker='o')
    plt.annotate('({}, {})'.format(round(vertex_x, 2), round(vertex_y, 2)), (vertex_x, vertex_y),
                 xytext=(0, 10), textcoords='offset points', ha='center')

    plt.legend()
    plt.show()

# 測試程式碼
plot_quadratic_equation(-3, 4, 7)
plot_quadratic_equation(5, -4, 9)
```
